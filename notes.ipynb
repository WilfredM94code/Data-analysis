```python
# There is a Pandas method used to get data from a list
# of tuples as shown below (or general iterables)

products = [('Product1','1$'),('Product2','$4')]
pd.DataFrame.from_records(products)
```

```python
# From every JSON file there are gonna be used only
# a few fields. To filter the required data from the
# JSON file there has to be provided a 'field list'

fields = ['id','all_artists','title','medium','acquisitionYear','height','width','units']

# Function to gather info from a JSON file
def records(file_path,fields):
    with open (file_path) as work_file:
        content = json.load(work_file)
    record = []
    for field in fields:
        record.append(content[field])
    return tuple(record)

# Function to search JSON files
def folder_walker(keys):
    json_root = os.path.join('resources','collection-master','artworks')
    artwork = []
    for root, _, files in os.walk(json_root):
        for file in files:
            if file.endswith('json'):
                record = records(os.path.join(root,file), keys)
                artwork.append (record)
                break
    df = pd.DataFrame.from_records(artwork, columns=keys,index='id')
    return df
```

```python
# Transformation

# Transformation occurs when a value is missing and is
# required to provide an assumption. This can be done by
# setting missing value to most frequent value or by 
# setting the mean as the missing value.

# To keep with this example the 'small_df' will get
# rows corrupted

small_df.loc[[19425,7265], 'medium'] = np.nan
# Two records corrupted
small_df.loc[[19425,7265], :]
# Explicit NaN display
```

```python
small_df
# Contextual NaN display
```

```python
# To change the NaN value for the most common on 
# the category there will be the next function

# Note that this is a particular case scenario

# For Series
def fill_values(series):
    values_counted = series.value_counts()
    if values_counted.empty:
        return series
    most_frequent = values_counted.index[0]
    new_medium = series.fillna(most_frequent)
    return new_medium

# For DataFrame
def transform_df(source_df):
    group_dfs = []
    for name, group_df in source_df.groupby ('artist'):
        filled_df = group_df.copy()
        filled_df.loc[:, "medium"] = fill_values (group_df ['medium'])
        group_dfs.append(filled_df)
    new_df = pd.concat(group_dfs)
    return new_df

new_df = transform_df(small_df)
new_df
# NaNs not present on DF
```

```python
new_df.loc[[19425,7265], :]
# No evidence of NaN
```

```python
# Groups operations - Built-in Methods BEST PRACTICE

# Aggregation Built-in (Numpy)
# This method work on Series/columns
aqc_years = small_df.groupby('artist')['acquisitionYear']
agg_bi_min = aqc_years.agg(np.min)
agg_bi_min
```

```python
# Aggregation Built-in (Pandas)
# This method work on Series/columns

agg_bi_min = aqc_years.min()
agg_bi_min
```

```python
# Transformation Built-in
# This method work on Series/columns

# Note that this method employs the function that that
# will rule the data transformation. This function will
# be the previous 'fill_values' function shown below.
# This function returns a unaltered serie if. This will 
# fill any NaN of the Serie with the most frequent. The
# syntax of the built-in group operation is shown below.
# This can work with lambda functions

def fill_values(series):
    values_counted = series.value_counts()
    if values_counted.empty:
        return series
    most_frequent = values_counted.index[0]
    new_medium = series.fillna(most_frequent)
    return new_medium

serie = small_df.groupby('artist')['medium']
serie = serie.transform(fill_values)
serie
```

```python
serie.loc[[19425,7265]]
# No NaN
```
