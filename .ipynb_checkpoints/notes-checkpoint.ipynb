```python
# To import JSON data there has to be clean data to
# work with. At the 'resources' folder there is a folder
# tree in which is stores multiple JSON files. These
# files are meant to be accessed to use them as JSON
# data within our code

import json
```

```python
# There is a Pandas method used to get data from a list
# of tuples as shown below (or general iterables)

products = [('Product1','1$'),('Product2','$4')]
pd.DataFrame.from_records(products)
```

```python
# From every JSON file there are gonna be used only
# a few fields. To filter the required data from the
# JSON file there has to be provided a 'field list'

fields = ['id','all_artists','title','medium','acquisitionYear','height','width','units']

# Function to gather info from a JSON file
def records(file_path,fields):
    with open (file_path) as work_file:
        content = json.load(work_file)
    record = []
    for field in fields:
        record.append(content[field])
    return tuple(record)

# Function to search JSON files
def folder_walker(keys):
    json_root = os.path.join('resources','collection-master','artworks')
    artwork = []
    for root, _, files in os.walk(json_root):
        for file in files:
            if file.endswith('json'):
                record = records(os.path.join(root,file), keys)
                artwork.append (record)
                break
    df = pd.DataFrame.from_records(artwork, columns=keys,index='id')
    return df
```

```python
# Transformation

# Transformation occurs when a value is missing and is
# required to provide an assumption. This can be done by
# setting missing value to most frequent value or by 
# setting the mean as the missing value.

# To keep with this example the 'small_df' will get
# rows corrupted

small_df.loc[[19425,7265], 'medium'] = np.nan
# Two records corrupted
small_df.loc[[19425,7265], :]
# Explicit NaN display
```

```python
small_df
# Contextual NaN display
```

```python
# To change the NaN value for the most common on 
# the category there will be the next function

# Note that this is a particular case scenario

# For Series
def fill_values(series):
    values_counted = series.value_counts()
    if values_counted.empty:
        return series
    most_frequent = values_counted.index[0]
    new_medium = series.fillna(most_frequent)
    return new_medium

# For DataFrame
def transform_df(source_df):
    group_dfs = []
    for name, group_df in source_df.groupby ('artist'):
        filled_df = group_df.copy()
        filled_df.loc[:, "medium"] = fill_values (group_df ['medium'])
        group_dfs.append(filled_df)
    new_df = pd.concat(group_dfs)
    return new_df

new_df = transform_df(small_df)
new_df
# NaNs not present on DF
```

```python
new_df.loc[[19425,7265], :]
# No evidence of NaN
```

```python
# Groups operations - Built-in Methods BEST PRACTICE

# Aggregation Built-in (Numpy)
# This method work on Series/columns
aqc_years = small_df.groupby('artist')['acquisitionYear']
agg_bi_min = aqc_years.agg(np.min)
agg_bi_min
```

```python
# Aggregation Built-in (Pandas)
# This method work on Series/columns

agg_bi_min = aqc_years.min()
agg_bi_min
```

```python
# Transformation Built-in
# This method work on Series/columns

# Note that this method employs the function that that
# will rule the data transformation. This function will
# be the previous 'fill_values' function shown below.
# This function returns a unaltered serie if. This will 
# fill any NaN of the Serie with the most frequent. The
# syntax of the built-in group operation is shown below.
# This can work with lambda functions

def fill_values(series):
    values_counted = series.value_counts()
    if values_counted.empty:
        return series
    most_frequent = values_counted.index[0]
    new_medium = series.fillna(most_frequent)
    return new_medium

serie = small_df.groupby('artist')['medium']
serie = serie.transform(fill_values)
serie
```

```python
serie.loc[[19425,7265]]
# No NaN
```

```python
# Filtering Built-in
# The syntax of the built-in group operation is shown 
# below. This can work with lambda functions

# Filtering Built-in
grouped_titles = small_df.groupby('artist')
# title_counts = grouped_titles.size().sort_values(ascending=False)
dup_titles_df = grouped_titles. filter (lambda x: len(x.index) > 15)
dup_titles_df.sort_values('title')
dup_titles_df
```

```python
# Iterating over arrays
#Iteration can occur over each element
it_array_a = np.arange(12)
for number in it_array_a:
    print (number)
```

```python
# Creating shallow copies 
# Shallow copies are copies of an object that refers to
# a same memory space. Which means that if one object
# is changed the every other object that makes reference
# to that object will access the modified value


shallow_copy_a = np.array(['Alamo','Barcelona','Colonia','Detroit'])
view_a = shallow_copy_a.view()
view_b = shallow_copy_a.view()

print (id(shallow_copy_a))
print (shallow_copy_a)
print (id(view_a))
print (view_a)
print (id(view_b))
print (view_b)
```

```python
# All these objects refer to a different memory space, 
# but Numpy considers each object as a shallow copy

view_a[0] = 'Aragua'

print (shallow_copy_a)
print (view_a)
print (view_b)
```

```python
# To check if a array is an shallow copy or a reference to
# an array there can be used the following attribute to 
# compare
view_a.base is shallow_copy_a
```

```python
# Deep copies
# To create a deep copy there can be used the 'copy()'
deep_copy_a = shallow_copy_a.copy()
print (deep_copy_a)
```

```python
# Note that deep copies are a resource meant to be used 
# wisely since it indiscriminate usage can flood RAM 
# memory
deep_copy_a[:], shallow_copy_a[:] = True , False
print (deep_copy_a)
print (shallow_copy_a)
```

```python
# Broadcasting

# Describe how NumPy treats arrays with different shapes
# during arithmetic operations.Subject to certain 
# constraints, the smaller array is "broadcast" across 
# the larger array so that they have compatible shapes.

# For instance, a 0D or 1x1 array (a plain number) can be 
# broadcasted over any nD array. This occurs since there's 
# a compatibility over the shape of every axis magnitude 
# and the dimention of the plain number

# This can also occur with a 3X6 array and a 3X1 array
# Same wouldn't be possible if there was an 3X2 array
# instead of the 3X1 array

# To broadcast an array over another there has to exists
# either concordance in between an array and the meant-
# to-be-broadcasted array in all the axis dimentions.
# The magnitude of the shape over one dimention can be
# either the value of the base array or 1 

 # This makes arrays compatibles

broadcasting_a = np.arange(50).reshape(5,5,2)
broadcasting_b = np.arange(10).reshape(1,5,2)
broadcasting_a + broadcasting_b
```

```python
broadcasting_a * broadcasting_b
```

```python
# Vectorization
# Numpy allows to execute functions item-wise over arrays
# by vectorizing them
def add_sub (a,b):
    if a > b:
        return a - b
    else:
        return a + b

print(add_sub(8,10))
print(add_sub(10,8)) 
```

```python
vec_a = np.vectorize(add_sub) # Vectorized function
vec_c = vec_a ([4,5,6,7,8],[8,7,6,5,4]) # Vectorized array
```
